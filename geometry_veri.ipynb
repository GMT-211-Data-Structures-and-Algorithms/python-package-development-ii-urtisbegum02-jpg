```python
import math

class Point:
    """
    2D düzlemde bir noktayı temsil eder.
    """
    def __init__(self, x, y, label=""):
        self.x = float(x)
        self.y = float(y)
        self.label = label

    def __repr__(self):
        return f"Point({self.label}: {self.x}, {self.y})"

    @classmethod
    def from_file_line(cls, line_str):
        """
        Dosyadaki '10,2, Bölüm Binası' formatını okur.
        """
        parts = line_str.strip().split(',')
        if len(parts) >= 3:
            return cls(parts[0], parts[1], parts[2].strip())
        return None

class Line:
    """
    Belirli bir isimle tanımlanan ve noktalardan oluşan bir çizgiyi temsil eder.
    """
    def __init__(self, name):
        self.name = name
        self.points = []

    def add_point(self, point):
        self.points.append(point)

    def calculate_length(self):
        """
        Çizginin toplam uzunluğunu hesaplar (Öklid mesafesi).
        """
        total_length = 0
        for i in range(len(self.points) - 1):
            p1, p2 = self.points[i], self.points[i+1]
            total_length += math.sqrt((p2.x - p1.x)**2 + (p2.y - p1.y)**2)
        return total_length

    def __repr__(self):
        return f"Line({self.name}, Points: {len(self.points)})"

# Örnek Kullanım (Test etmek istersen)
if __name__ == "__main__":
    p1 = Point(0, 0, "Başlangıç")
    p2 = Point(3, 4, "Bitiş")
    yol = Line("Ana Yol")
    yol.add_point(p1)
    yol.add_point(p2)
    print(f"{yol.name} Uzunluğu: {yol.calculate_length()}")
```
